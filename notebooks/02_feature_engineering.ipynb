### Feature Engineering
#### Objective:
#### Generate lag features, rolling statistics, and time-based features to capture temporal sales patterns.

```python
import pandas as pd
import numpy as np

# Load cleaned data
df = pd.read_csv(r"data/processed/cleaned_sales.csv")
df["date"] = pd.to_datetime(df["date"])

# Log transform sales (to handle skewness)
df["sales_log"] = np.log1p(df["sales"])

df["week_of_year"] = df["date"].dt.isocalendar().week.astype(int)
df["month"] = df["date"].dt.month


# Lag features
df["lag_1"] = df.groupby(["store", "item"])["sales_log"].shift(1)
df["lag_7"] = df.groupby(["store", "item"])["sales_log"].shift(7)
df["lag_14"] = df.groupby(["store", "item"])["sales_log"].shift(14)

# Rolling features
df["roll_mean_7"] = (
    df.groupby(["store","item"])["sales_log"]
      .transform(lambda x: x.shift(1).rolling(7).mean())
)

df["roll_std_7"] = (
    df.groupby(["store","item"])["sales_log"]
      .transform(lambda x: x.shift(1).rolling(7).std())
)

df["roll_mean_14"] = (
    df.groupby(["store","item"])["sales_log"]
      .transform(lambda x: x.shift(1).rolling(14).mean())
)

df["roll_mean_28"] = (
    df.groupby(["store","item"])["sales_log"]
      .transform(lambda x: x.shift(1).rolling(28).mean())
)

# Drop missing values created by lag/rolling
df = df.dropna()

# Save the feature dataset
df.to_csv(r"data/processed/features.csv", index=False)
df.head()
```
